# Normalized Laplacian Tests

Note here we had to make some changes to the code to be able to get $\mathbf{u}$ in the span of the correct $\chi_k$'s

## Regime 1

$$\epsilon = \tau^2, \gamma = \tau^{\alpha}$$

This corresponds to ``little_oh = False``

```python
rand = False
Ns = [100,100,100]
fs = 5
little_oh = False

Ns, u, B, labeled, H, draws, W_0 = overall_setup_norm(rand, Ns, fs)
param_str = 'rand_%d_Ns_%d_%d_%d_fs_%d_lo_%d' % (int(rand), Ns[0], Ns[1], Ns[2], fs, int(little_oh))

n = 20
T   = np.power(0.6, range(5, n+5))
ALPHAS = [0.5, 1, 1.5, 2, 2.5, 3.]
TRC, TRCBC, normC, normC_tau_inv_u, normC_tau_inv_u_p1, normC_tau_inv_u_p2, BIAS = run_test_norm(T, 
                                                            ALPHAS, Ns, B, labeled, u, W_0, little_oh=False)    

_save = False
print(param_str)
plot_trC(T, TRC, ALPHAS, param_str, save=_save, Jval=6)
plot_trCBC(T, TRCBC, ALPHAS, param_str, save=_save, Jval=8)
plot_data(T, BIAS, ALPHAS, param_str, title_= r'$|C^*C_{\tau, \epsilon}^{-1}\mathbf{v}|$', save=False, Jval=-1)
plot_data(T, normC*normC_tau_inv_u, ALPHAS, param_str,  title_=r'$\|C^*\|_2|C_{\tau, \epsilon}^{-1}\mathbf{v}|$', save=False, Jval=-1)
plot_data(T, TRC*normC_tau_inv_u, ALPHAS, param_str,  title_=r'$\mathrm{Tr}(C^*)|C_{\tau, \epsilon}^{-1}\mathbf{v}|$', save=False, Jval=-1)
plot_data(T, normC, ALPHAS, param_str, title_=r'$\|C^*\|_2$', save=False, Jval=-1)
plot_data(T, normC_tau_inv_u, ALPHAS, param_str,  title_=r'$|C_{\tau, \epsilon}^{-1}\mathbf{v}|$', save=False, Jval=-1)
plot_data(T, normC_tau_inv_u_p1+normC_tau_inv_u_p2, ALPHAS, param_str,  
          title_=r'$|C_{\tau, \epsilon}^{-1}P_\epsilon\mathbf{v}| + |C_{\tau, \epsilon}^{-1}(I - P_\epsilon)\mathbf{v}|$', save=False, Jval=-1)
plot_data(T, normC_tau_inv_u_p1, ALPHAS, param_str,  
          title_=r'$|C_{\tau, \epsilon}^{-1}P_\epsilon\mathbf{v}|$', save=False, Jval=-1)
plot_data(T, normC_tau_inv_u_p2, ALPHAS, param_str,  
          title_=r'$|C_{\tau, \epsilon}^{-1}(I - P_\epsilon)\mathbf{v}|$', save=False, Jval=-1)
```
